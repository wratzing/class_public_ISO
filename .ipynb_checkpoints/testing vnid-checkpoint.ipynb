```python
cosmo1=Class()
cosmo1.set({'omega_b':0.0243828,'omega_cdm':0.1201075,'T_cmb':2.7255,'h':0.67810,'tau_reio':0.05430842,
            'N_ur':3.0,'N_idr':1.0,'idr_nature':'fluid'})
cosmo1.set({'output':'mTk,dTk,tCl','k_output_values':10,'ic':'ad,vid','vid_phi':0.1,'vid_cos_theta'})

tstart=time.perf_counter()
cosmo1.compute()
print(time.perf_counter()-tstart)
```
